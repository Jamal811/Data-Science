```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage to scrape
url = 'https://ced.neduet.edu.pk/node/207'

# Send an HTTP request to the URL
response = requests.get(url)

# Parse the content of the request with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# List to hold the extracted data
faculty_details = []

# Scrape faculty details
for faculty in soup.find_all('div', class_='team-single-page'):
    name_tag = faculty.find('h1', id='')
    role_tag = faculty.find('div', class_='job')
    
    email_tags = soup.find('div', class_='field field--name-field-team-contact field--type-string field--label-hidden field__items').find_all('div', class_='field__item')
    email_tag = email_tags[1] if len(email_tags) > 1 else None

    name = name_tag.text.strip() if name_tag else 'N/A'
    role = role_tag.text.strip() if role_tag else 'N/A'
    email = email_tag.text.strip() if email_tag else 'N/A'

    

    
    faculty_details.append({
        'name': name,
        'role': role,
        'email': email,
    })

# Create a DataFrame
df_faculty = pd.DataFrame(faculty_details)

# Print the DataFrame
df_faculty
```
