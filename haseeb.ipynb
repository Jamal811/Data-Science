```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data based on the provided table
data = {
    'Model': ['BiLSTM', 'BiGRU', 'ConvBiLSTM', 'ConvBiGRU', '1D-CNN'] * 2,
    'Word Embedding': ['Word2Vec'] * 5 + ['FastText'] * 5,
    'Accuracy': [85.98, 86.73, 87.23, 85.57, 87.47, 86.35, 85.92, 86.92, 87.17, 86.47]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set style
sns.set(style='whitegrid')

# Plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Model', y='Accuracy', hue='Word Embedding', data=df, palette=['#1f77b4', '#ff7f0e'])
plt.title('Comparing Models Accuracy Using Different Word Embedding Methods on the IMDB Dataset')
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.xticks(rotation=45)
plt.legend(title='Word Embedding', loc='lower center', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Adding accuracy labels on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', ((p.get_x() + p.get_width() / 2), p.get_height()), ha='center', va='bottom')

plt.show()


```
