```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin

# Main URL
url = 'https://ced.neduet.edu.pk/'

# Send a GET request to the main page
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# List to hold faculty details
faculty_details = []

# Locate the elements containing the profile page links
profile_links = [urljoin(url, link['href']) for link in soup.select('.team-name a')]

# Visit each profile page and scrape data
for profile_link in profile_links:
    profile_response = requests.get(profile_link)
    profile_soup = BeautifulSoup(profile_response.content, 'html.parser')
    
    name_tag = profile_soup.find('h1', class_='')
    role_tag = profile_soup.find('div', class_='job')
    email_tags = profile_soup.find('div', class_='field field--name-field-team-contact field--type-string field--label-hidden field__items').find_all('div', class_='field__item')
    email_tag = email_tags[1] if len(email_tags) > 1 else None

    name = name_tag.text.strip() if name_tag else 'N/A'
    role = role_tag.text.strip() if role_tag else 'N/A'
    email = email_tag.text.strip() if email_tag else 'N/A'
    
    # Append data to faculty_details list
    faculty_details.append({
        'Name': name,
        'Designation': role,
        'Email': email,
    })

# Create a DataFrame
Civil_faculty = pd.DataFrame(faculty_details)
```
