```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
url = 'https://cusit.edu.pk/cusitnew/arch/faculty.php'  # Replace with the actual URL
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.findAll('h1')
```

```python
soup.findAll('h1')[0].text.strip()
```

```python
# Department name
department_name = "Department of Architecture"  # Replace with actual department name if it can be extracted

# List to hold the extracted data
faculty_details = []

# Scrape faculty details
for faculty in soup.find_all('div', class_='card-body pt-6 text-center'):
    name_tag = faculty.find('h4', class_='color-primary')
    p_tags = faculty.find_all('p')
    
    name = name_tag.text.strip() if name_tag else 'N/A'
    role = p_tags[0].text.strip() if len(p_tags) > 0 else 'N/A'
    email = p_tags[1].text.strip() if len(p_tags) > 1 else 'N/A'
    
    faculty_details.append({
        'name': name,
        'role': role,
        'email': email,
        'department': department_name
    })

# Create a DataFrame
df_architecture = pd.DataFrame(faculty_details)

# Print the DataFrame
df_architecture
```
