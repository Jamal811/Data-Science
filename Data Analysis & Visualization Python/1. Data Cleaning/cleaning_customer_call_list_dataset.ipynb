```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
#remove extra things from phone number column
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '')
df
```

```python
#convert the column to string all values, after that we can do formating in next step bacause formating works on string value
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
#applying formating on phone number column
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
#remove na-- and Na-- from the column
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
df
```

```python
#splitting  columns of address
df[["Street_Address", "State", "Zip_Code"]]= df["Address"].str.split(',', 2, expand=True)
df
```

```python
# convert all the yes to Y and no to N in the Paying Customer column
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
# convert all the yes to Y and no to N in the "Do_Not_Contact" column
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
#Remove N/a values from dataset
df = df.replace('N/a', '')
df
```

```python
#Remove NaN values from dataset
df = df.fillna('')
df
```

```python
#drop those rows which is Y in "Do_Not_Contact"
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Y':
    df.drop(x, inplace = True)

df
#2nd method of doing this
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
#drop those rows which has no contact number in "Phone_Number"
for x in df.index:
  if df.loc[x, "Phone_Number"] == '':
    df.drop(x, inplace = True)

df
```

```python
#reset the index
df = df.reset_index(drop=True)
df
```

```python
#remove address column
df = df.drop(columns = "Address")
df
```
