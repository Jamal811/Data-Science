```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Expanded_data_with_more_features.csv')
df.head()
```

## Drop unnamed column

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
# df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('05-Oct', '5 - 10')
```

## Gender distrribution

```python
plt.figure(figsize=(5,4))
gen_plot = sns.countplot(x='Gender',data=df)
gen_plot.bar_label(gen_plot.containers[0]) # add values total count on the top
plt.title("Gender Disttribution")

plt.show()
```

```python
parent_edu = df.groupby("ParentEduc").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(parent_edu)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(parent_edu, annot=True)
plt.title("The Impact of Parent's Education on Student's Score")

plt.show()
```

```python
parent_martial = df.groupby("ParentMaritalStatus").agg({"MathScore": "mean", "ReadingScore": "mean", "WritingScore": "mean"})
print(parent_martial)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(parent_martial, annot=True)
plt.title("The Impact of Martial Status of Parents on Student's Score")
plt.show()
```

## Distribution of Ethinic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

my_list = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup'],]
print(my_list)
labels =["group A", "group B", "group C", "group D", "group E"] 

plt.pie(my_list, labels=labels, autopct="%1.2f%%")
plt.title("Distribution of Ethinic Group")
plt.show()
```
