## Heatmaps

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, linewidths=0.2)
```

```python
sns.heatmap(x, linewidths=0.2, linecolor='g', annot=True, fmt='d')
```

```python
sns.heatmap(x, cmap='viridis')
```

```python
sns.heatmap(x, cmap='summer')
```

```python
sns.heatmap(x, cmap='winter')
```

## Cluster Map

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x, z_score=0)
```

```python
sns.clustermap(x, z_score=0, annot=True) 
```

```python
sns.clustermap(x, z_score=0, annot=True, col_cluster=False) 
```

## distplots

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'], kde=False)
```

```python
sns.distplot(titanic['age'], hist=False)
```

```python
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
```

```python
sns.distplot(titanic[titanic['survived'] ==1]['age'])
sns.distplot(titanic[titanic['survived'] ==0]['age'])
```
